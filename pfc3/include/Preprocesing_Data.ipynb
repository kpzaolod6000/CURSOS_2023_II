### Estableciendo separador para cada token

```python
def textProcessingToString(row):
  strText = ""
  for i in range(len(row)):
    # print(text)
    if (i == len(row) -1):
      strText = strText + row[i]
    else:
      strText = strText + row[i] + "##"
  return strText
```

### Stemming and Stop Word

```python
import os

def setWordsInTxt(arr,nameFile):
    
    archivo_path = os.path.join("data/" + nameFile, f"{nameFile}.txt")
    with open(archivo_path, "w") as file:
        for wordsByDoc in arr:
            for word in wordsByDoc:
                file.write(word + "\n")
            file.write("##\n")
```

### Keywords

```python
def textProcessingKeywords(df_data):
    keywordsPre_lem = []
    keywordsPre_stem = []
    
    pattern = r'^[^a-zA-Z]*|[^a-zA-Z0-9]*$'
    pattern_number = r'\b\d+\b'
    
    i_ = 1
    for index, row in df_data.iterrows():

        ## TO KEYWORDS

        _keywords = row['Keywords']
        cleaned_text = _keywords.lower()
        cleaned_text = cleaned_text.replace(",", " ")
        
        # Tokenización
        tokens = word_tokenize(cleaned_text)  #Tokenización
        tokens = [re.sub(pattern_number, '', token) for token in tokens] #eliminado numeros
        tokens = [re.sub(pattern, '', token) for token in tokens] #eliminando caracteres especiales prefijo y sufijos

        print(i_)
        print(_keywords)
        print(tokens)


        # Eliminación de stopwords
        stop_words = set(stopwords.words('english'))  # Lista de stopwords en español
        filtered_tokens = [token for token in tokens if token not in stop_words]
        # print(filtered_tokens)

        # cleaned_words = [re.sub(r'[^\w\s-]', '', word) for word in filtered_tokens]
        cleaned_words = [word for word in filtered_tokens if len(word) > 1 ]
        cleaned_words = list(set(cleaned_words))

        # print("cleaned word keyword", cleaned_words)

        ## Stemming y lematización
        stemmer_k = PorterStemmer()
        lemmatizer_k = WordNetLemmatizer()
        stemmed_tokens_k = [stemmer_k.stem(token) for token in cleaned_words]
        print("stemmed word keyword", stemmed_tokens_k)
        lemmatized_tokens_k = [lemmatizer_k.lemmatize(token) for token in cleaned_words]
        print("lemmatized word keyword",lemmatized_tokens_k)

        keywordsPre_lem.append(lemmatized_tokens_k)
        keywordsPre_stem.append(stemmed_tokens_k)
        i_+=1
        # dataPre = lemmatized_tokens_t + lemmatized_tokens_k

    return keywordsPre_lem, keywordsPre_stem
```

### Title

```python
def textProcessingTitle(df_data):

    titlePre_lem = []
    titlePre_stem = []
    
    pattern = r'^[^a-zA-Z]*|[^a-zA-Z0-9]*$'
    pattern_number = r'\b\d+\b'
    i_ = 0    
    for index, row in df_data.iterrows():

        ## TO TITLE

        _title = row['Title'].strip()

        cleaned_text = _title.lower()  # Convertir a minúsculas

        # Tokenización
        tokens = word_tokenize(cleaned_text)  #Tokenización
        tokens = [re.sub(pattern_number, '', token) for token in tokens] #eliminado numeros
        tokens = [re.sub(pattern, '', token) for token in tokens] #eliminando caracteres especiales prefijo y sufijos

        print(i_)
        print(_title)
        print(tokens)


        # Eliminación de stopwords
        stop_words = set(stopwords.words('english'))  # Lista de stopwords en español
        filtered_tokens = [token for token in tokens if token not in stop_words]
        # print(filtered_tokens)

        # Eliminar signos de puntuación y la parte "-based" usando expresiones regulares
        # cleaned_words = [re.sub(r'[^\w\s-]', '', word) for word in filtered_tokens]
        cleaned_words = [word for word in filtered_tokens if len(word) > 1]
        cleaned_words = list(set(cleaned_words))
        # print("cleaned word", cleaned_words)

        
        ## Stemming y lematización
        stemmer_t = PorterStemmer()
        lemmatizer_t = WordNetLemmatizer()
        stemmed_tokens_t = [stemmer_t.stem(token) for token in cleaned_words]
        print("stemmed word keyword", stemmed_tokens_t)
        lemmatized_tokens_t = [lemmatizer_t.lemmatize(token) for token in cleaned_words]
        print("lemmatized word keyword",lemmatized_tokens_t)

        titlePre_lem.append(lemmatized_tokens_t)
        titlePre_stem.append(stemmed_tokens_t)

        i_+=1

    return titlePre_lem, titlePre_stem
```

## Cleaned Text of Specials Characters

```python
def hasSpecialChar(word):
    special_characters = "!@#$%^&*()_+-=[]{}|;:'\",.<>?/"
    count = 0
    for char in special_characters:
        if char in word:
            count += 1
            if count >= 2:
                return True
    return False

def cleanWordsByDoc(namePath,nameFileInput,nameFileOutput,wordDeleted):

    input_file_name = os.path.join("data/"+namePath , nameFileInput+".txt")
    output_file_name = os.path.join("data/"+namePath, nameFileOutput+".txt")
    output_file_deleted_name = os.path.join("data/"+namePath, wordDeleted+".txt")

    with open(input_file_name, "r") as input_file, open(output_file_name, "w") as output_file, open(output_file_deleted_name, "w") as output_wordDe_file:
        for line in input_file:
            words = line.split()
            for word in words:
                if (word != "##"):
                    if hasSpecialChar(word):
                        response = input(f"The word '{word}' contains special characters. Do you want to include it in the output file? (y/n): ")
                        if response.lower() == 'y':
                            output_file.write(word + "\n")
                            continue
                        else:
                            output_wordDe_file.write(word + "\n")
                            continue
                    output_file.write(word + "\n")
                else:
                    output_file.write(word + "\n")
```

### Delete repeated words per document

```python
def getWordsOfTxt(namePath,nameFile):

    input_file_name = os.path.join("data/"+namePath, nameFile+".txt")

    arrTotal = []
    arrCurrent = []
    with open(input_file_name, "r") as input_file:
        
        for linea in input_file:
            word = linea.strip()
            # print(word)
            if word == '##':
                arrTotal.append(arrCurrent)
                arrCurrent = []
            else:
                arrCurrent.append(word)
    return arrTotal

```

```python
def removingRepeatedWords(namePath,nameFile,nameFileDestiny):
    arr_aux = getWordsOfTxt(namePath,nameFile)
    print(len(arr_aux))
    output_file = os.path.join("data/" + namePath, f"{nameFileDestiny}.txt")
    with open(output_file, "w") as file:
        for docText in arr_aux:
            Deletedtext = list(set(docText))    
            
            for word in Deletedtext:
                file.write(word + "\n")
            file.write("##\n")
```
